```python
import numpy as np
import matplotlib.pyplot as plt

fig_name_suffix_ = 'learn_plot_'
```

```python
x_list = [1,2,3,4,5]
y_list = [1,1,2,3,5]

plt.plot(x_list, y_list, linewidth=2.0)
plt.xlim([np.min(x_list)*1.05, np.max(x_list)*1.05])
plt.ylim([np.min(y_list), np.max(y_list)])

plt.title("Title")
plt.ylabel(r'Y', fontsize=18)
plt.xlabel(r'X', fontsize=18)
plt.savefig(f"figure/{fig_name_suffix_}list-list.png")
plt.show()
```

```python
graph_line = 4
x_list = [-3,-2,-1,0,1,2,3]


fig, ax = plt.subplots(figsize=(10, 7))

for i in range(graph_line):
    y_list = [x^i for x in x_list] if (i==0) else [1 for x in x_list]
    ax.plot(x_list, y_list, linewidth=2.0)


ax.legend()
# ax.set_xlim(min(x_list), max(x_list))
# ax.set_ylim(min(y_list)*1.05, max(y_list)*1.05)

plt.title("Title")
plt.ylabel(r'Y', fontsize=18)
plt.xlabel(r'X', fontsize=18)

ax.set_xticks([min(x_list),0,max(x_list)])
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)

plt.savefig(f"figure/{fig_name_suffix_}multiple-graph.png")
plt.show()
```
